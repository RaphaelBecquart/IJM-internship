```python
from glob import glob
import json
import matplotlib.pyplot as plt

import numpy as np
```

```python
file_list = glob('./data_json/*.json')
```

```python
angle_list = []
for a_file in file_list:
    with open(a_file) as json_file:
        data = json.load(json_file)
        
    key = a_file[a_file.rfind('/')+1:a_file.rfind('.')]
    
    try:
        tmp_angle = [data[key][idx]['angle'] for idx in range(len(data[key]))]
        angle_list += tmp_angle
    except:
        pass
```

```python
angle_list = [angle for el in sublist for sublist in angle_list]
```

```python
plt.figure(figsize=(20,5))
plt.hist(angle_list, bins = np.arange(0, 361, 5), histtype = 'stepfilled')
plt.xticks(np.arange(0, 361, 45), size = 14)
plt.yticks(size = 14)
plt.xlabel('angle (in degrees)', size = 20)
plt.ylabel('number of centrioles', size = 20)
plt.xlim(0,360)
plt.ylim(0, 22000)
plt.title('repartition of centriole orientation', size = 26)
```
